**The ADI (Alternating Direction Implicit) method**

The ADI (Alternating Direction Implicit ) algorithm is a simple generalisation of the 1D implicit algorithm to solve the linear diffusion equation to 2D and 3D. Assuming that the problem has been discretised on a mesh of dimension $n_x\times n_y$, it consists in solving $n_y$ 1D problems of size $n_x$ followed by solving $n_x$ 1D problems of size $n_y$.

First solve:

$h_{i,j}^{k+1/2}=h_{i,j}^k+\frac{K\Delta t}{2\Delta x^2}(h_{i+1,j}^{k+1/2}-2h_{i,j}^{k+1/2}+h_{i-1.j}^{k+1/2})+\frac{K\Delta t}{2\Delta y^2}(h_{i,j+1}^k-2h_{i,j}^k+h_{i,j-1}^k)$

then solve:

$h_{i,j}^{k+1}=h_{i,j}^{k+1/2}+\frac{K\Delta t}{2\Delta x^2}(h_{i+1,j}^{k+1/2}-2h_{i,j}^{k+1/2}+h_{i-1.j}^{k+1/2})+\frac{K\Delta t}{2\Delta y^2}(h_{i,j+1}^{k+1}-2h_{i,j}^{k+1}+h_{i,j-1}^{k+1})$

For ease of implementation, one first writes a routine/function solving implicitly a 1D problem by solving a tridiagonal matrix (as done earlier in the course). One then writes a series of loops calling the function $n_x$ then $n_y$ times. The algorithm remains of complexity $O(n_x\times n_y)$.

```python
@numba.njit
## Tri Diagonal Matrix Algorithm(a.k.a Thomas algorithm) solver
def TDMAsolver(a, b, c, d):
    '''
    TDMA solver, a b c d can be NumPy array type or Python list type.
    refer to http://en.wikipedia.org/wiki/Tridiagonal_matrix_algorithm
    and to http://www.cfd-online.com/Wiki/Tridiagonal_matrix_algorithm_-_TDMA_(Thomas_algorithm)
    a is lower diagonal
    b is diagonal
    c is upper diagonal
    d is rhs
    '''
    nf = len(d) # number of equations
    ac = a.copy()
    bc = b.copy()
    cc = c.copy()
    dc = d.copy()
    for it in range(1, nf):
        mc = ac[it-1]/bc[it-1]
        bc[it] = bc[it] - mc*cc[it-1] 
        dc[it] = dc[it] - mc*dc[it-1]
        	    
    xc = bc
    xc[-1] = dc[-1]/bc[-1]

    for il in range(nf-2, -1, -1):
        xc[il] = (dc[il]-cc[il]*xc[il+1])/bc[il]

    return xc
```

Let's create a 2D problem

```python
Lx = 1000 # the length of the problem in the x-direction
Ly = 1000 # the length of the problem in the y-direction
nx = 101 # number of nodes in the x-direction
ny = 101 # number of nodes in the y-direction
dx = Lx/(nx - 1) # spatial discretization in x
dy = Ly/(ny - 1) # spatial discretization in y
x = np.linspace(0, Lx, nx) # 1D x-coordinate
y = np.linspace(0, Ly, ny) # 1D y-coordinate
X, Y = np.meshgrid(x,y) # 2D X and Y coordinates

h0 = 100 # amplitude of initial topography
n = 2 # number of periods in x-direction
m = 3 # number of periods in y-direction
h = h0*np.sin(2*np.pi*X*n/Lx)*np.sin(2*np.pi*Y*m/Ly) # initial topography

K = 1 # transport coeeficient or difusivity
tfinal = 10000 # total duration of model run
nt = 500 # number of times steps
dt = tfinal/nt # time step

print('Courant:', dt/dx**2*K, dt/dy**2*K) # print Courant condition
```

We now develop a 2D version of the ADI solver

```python
def solve2D(h, Fx, Fy):
    
    nx, ny = h.shape
    
    diag = np.ones(nx)*(1 + 2*Fx/2)
    lower = np.ones(nx-1)*(-Fx/2)
    upper = lower.copy()
    diag[0] = 1
    upper[0] = 0
    diag[-1] = 1
    lower[-1] = 0
    hhalf = h.copy()
    for j in range(1,ny-1):
        rhs = h[:,j].copy()
        rhs[1:-1] = rhs[1:-1] + Fy/2*(h[1:-1,j+1] - 2*h[1:-1,j] + h[1:-1,j-1])
        hhalf[:,j] = TDMAsolver(lower, diag, upper, rhs)
 
    diag = np.ones(ny)*(1 + 2*Fy/2)
    lower = np.ones(ny-1)*(-Fy/2)
    upper = lower.copy()
    diag[0] = 1
    upper[0] = 0
    diag[-1] = 1
    lower[-1] = 0
    for i in range(1,nx-1):
        rhs = hhalf[i,:].copy()
        rhs[1:-1] = rhs[1:-1] + Fx/2*(hhalf[i+1,1:-1] - 2*hhalf[i,1:-1] + hhalf[i-1,1:-1])
        h[i,:] = TDMAsolver(lower, diag, upper, rhs)

    return h

```

and use it to solve the problem

```python
h = h0*np.sin(2*np.pi*X*n/Lx)*np.sin(2*np.pi*Y*m/Ly) # initial topography

Fx = dt*K/dx**2
Fy = dt*K/dy**2
for step in range(nt):
    h = solve2D(h, Fx, Fy)

taux = Lx**2/4/np.pi**2/n**2/K
tauy = Ly**2/4/np.pi**2/m**2/K
ha = h0*np.exp(-tfinal/taux)*np.sin(2*np.pi*X*n/Lx)*np.exp(-tfinal/tauy)*np.sin(2*np.pi*Y*m/Ly)

```

Let's compute the error

```python
err = np.abs(h-ha).max()/h0
print(err)
```

and plot the numerical solution, the analytical solution and the difference

```python
fig, ax = plt.subplots(ncols=3, figsize=(15,4))
cp0 = ax[0].contourf(h)
cp1 = ax[1].contourf(ha)
cp2 = ax[2].contourf(h-ha)

plt.colorbar(cp0,ax=ax[0])
plt.colorbar(cp1,ax=ax[1])
plt.colorbar(cp2,ax=ax[2])

ax[0].set_title('ADI numerical')
ax[1].set_title('analytical')
ax[2].set_title('difference')
```
