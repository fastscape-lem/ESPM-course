```python
import numpy as np
import matplotlib.pyplot as plt
import numba
```

```python
import TDMAsolver
```

```python
Lx = 1000 # the length of the problem in the x-direction
Ly = 1000 # the length of the problem in the y-direction
nx = 101 # number of nodes in the x-direction
ny = 101 # number of nodes in the y-direction
dx = Lx/(nx - 1) # spatial discretization in x
dy = Ly/(ny - 1) # spatial discretization in y
x = np.linspace(0, Lx, nx) # 1D x-coordinate
y = np.linspace(0, Ly, ny) # 1D y-coordinate
X, Y = np.meshgrid(x,y) # 2D X and Y coordinates

h0 = 100 # amplitude of initial topography
n = 2 # number of periods in x-direction
m = 3 # number of periods in y-direction
h = h0*np.sin(2*np.pi*X*n/Lx)*np.sin(2*np.pi*Y*m/Ly) # initial topography

K = 1 # transport coeeficient or difusivity
tfinal = 10000 # total duration of model run
nt = 500 # number of times steps
dt = tfinal/nt # time step

print('Courant:', dt/dx**2*K, dt/dy**2*K) # print Courant condition
```

```python
cp = plt.contourf(h)
plt.colorbar(cp)
```

$$h_{i,j}^{k+1/2}=h_{i,j}^k+\frac{K\Delta t}{2\Delta x^2}(h_{i+1,j}^{k+1/2}-2h_{i,j}^{k+1/2}+h_{i-1.j}^{k+1/2})+\frac{K\Delta t}{2\Delta y^2}(h_{i,j+1}^k-2h_{i,j}^k+h_{i,j-1}^k)$$

$$h_{i,j}^{k+1}=h_{i,j}^{k+1/2}+\frac{K\Delta t}{2\Delta x^2}(h_{i+1,j}^{k+1/2}-2h_{i,j}^{k+1/2}+h_{i-1.j}^{k+1/2})+\frac{K\Delta t}{2\Delta y^2}(h_{i,j+1}^{k+1}-2h_{i,j}^{k+1}+h_{i,j-1}^{k+1})$$

```python
def solve2D(h, Fx, Fy):
    
    nx, ny = h.shape
    
    diag = np.ones(nx)*(1 + 2*Fx/2)
    lower = np.ones(nx-1)*(-Fx/2)
    upper = lower.copy()
    diag[0] = 1
    upper[0] = 0
    diag[-1] = 1
    lower[-1] = 0
    hhalf = h.copy()
    for j in range(1,ny-1):
#        rhs = h[:,j]
#        rhs[1:-1] = rhs[1:-1] + Fy/2*(h[1:-1,j+1] - 2*h[1:-1,j] + h[1:-1,j-1])
        rhs = h[:,j] + Fy/2*(h[:,j+1] - 2*h[:,j] + h[:,j-1])
        hhalf[:,j] = TDMAsolver.TDMAsolver(lower, diag, upper, rhs)
 
    diag = np.ones(ny)*(1 + 2*Fy/2)
    lower = np.ones(ny-1)*(-Fy/2)
    upper = lower.copy()
    diag[0] = 1
    upper[0] = 0
    diag[-1] = 1
    lower[-1] = 0
    for i in range(1,nx-1):
#        rhs = hhalf[i,:]
#        rhs[1:-1] = rhs[1:-1] + Fx/2*(hhalf[i+1,1:-1] - 2*hhalf[i,1:-1] + hhalf[i-1,1:-1])
        rhs = hhalf[i,:] + Fx/2*(hhalf[i+1,:] - 2*hhalf[i,:] + hhalf[i-1,:])
        h[i,:] = TDMAsolver.TDMAsolver(lower, diag, upper, rhs)

    return h
```

```python
h = h0*np.sin(2*np.pi*X*n/Lx)*np.sin(2*np.pi*Y*m/Ly) # initial topography

Fx = dt*K/dx**2
Fy = dt*K/dy**2
for step in range(nt):
    h = solve2D(h, Fx, Fy)

taux = Lx**2/4/np.pi**2/n**2/K
tauy = Ly**2/4/np.pi**2/m**2/K
ha = h0*np.exp(-tfinal/taux)*np.sin(2*np.pi*X*n/Lx)*np.exp(-tfinal/tauy)*np.sin(2*np.pi*Y*m/Ly)
```

```python
err = np.abs(h-ha).max()/h0
print(err)
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(15,4))
cp0 = ax[0].contourf(h)
cp1 = ax[1].contourf(ha)
cp2 = ax[2].contourf(h-ha)

plt.colorbar(cp0,ax=ax[0])
plt.colorbar(cp1,ax=ax[1])
plt.colorbar(cp2,ax=ax[2])

ax[0].set_title('ADI numerical')
ax[1].set_title('analytical')
ax[2].set_title('difference')
```
