# Numpy

*Numpy* is a package for scientific computing based on the manipulation of arrays

```python
import numpy as np
```

```python
a = np.array([0,1,2,3,4])
print(a)
print(type(a))
a
```

```python
a = np.ones(10)
print(a)
```

```python
a = np.linspace(0,10,5)
print(a)
```

```python
a = np.arange(0,10,2)
print(a)
```

### Array types

Numpy array contains numbers that can be of many types, including int, float, complex (the default is float)

```python
x = np.ones(10, int)
print(x)
```

### Multi-dimensional arrays

Arrays can have more than one dimensions

```python
a = np.array([[0,0,0],[1,1,1]])
a
```

```python
a = np.ones((3,2))
a
```

```python
x = np.linspace(0,10,11)
print(x)
y = np.linspace(0,5,6)
print(y)
X, Y = np.meshgrid(x,y)
print(X)
print(Y)
```

Multi-dimensional arrays are stored in row order and accessed through multiple indeces

```python
print(X[0,1])
print(X[0][1])
```

```python
print(X[0,9])
```

```python
print(X[:,::-1])
```

### Vector operations

*Numpy*'s strength comes from the efficiency of vector operations

```python
a = np.ones(5)
a = a*2
print(a)
```

```python
a= a + 0.3*a
print(a)
```

Although it is sometimes compulsory to use loops to perform an operation on the elements of an array, it is always much preferable to use a vector operation vecause it is optimized

However, operations may need to be performed in a given order, for example, which implies the use of a loop

```python
a = np.ones(10)
a += 1
print(a)

b = np.ones(10)
b[1:] = b[1:]+b[:-1]
print(b)

c = np.ones(10)
for i in range(1,len(c)):
    c[i] += c[i-1]
print(c)

```

### Numerical computational tools

*Numpy* offers a wide range of mathematical functions and operations

```python
a = np.sqrt(a)
print(a)
```

```python
print(np.sin(a))
print(np.log10(a+1))
print(np.pi)
```

```python
a = np.linspace(0,np.pi,11)
print(a)
print(np.where(a>2,a,-a))
```
