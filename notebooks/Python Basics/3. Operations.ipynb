# Operations

### Main operations

Operations include:
1. "=" assignment
2. "+" addition
3. "-" substraction
4. "*" multiplication
5. "/" division
6. "%" modulo (remainder of integer division)
7. "//" division and rounding down to integer value
8. "**" power

```python
a = 20
print(a+1)
print(a-3)
print(a*4)
print(a/7)
print(a%3)
print(a//3)
print(a**2)
```

### Abbreviations

Operations can also be abbreviated by combining "=" and the result replaces the original variable:

for example, a += 1 is equivalent to a = a + 1


```python
a = 3
a += 1
print(a)
```

a %= 2 is equivalent to a = a%2

```python
a = 7
a %= 2
print(a)
```
