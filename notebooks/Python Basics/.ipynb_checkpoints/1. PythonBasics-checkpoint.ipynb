# Python Basics #

Overview of essential Python for the Earth Surface Process Modeling course

*Python* is an interpreted, object-oriented language that is now broadly used for scientific computing

It is a high level language used for scripting with a rich library support

It is free, platform independent and designed to be compact

*Python* is also dynamically typed, which means that variable type is decided at run time (no need to declare variables)

```python
a = 1
another_number = 2
sum = a + another_number
print(sum)
```

### Multi-line commands:

Sometimes a line of code is too long and neeeds to be split between several lines; For this use brackets ()

```python
a = 1 + 2 + 3 + 4
a = (1 +
    2 +
    3 +
    4)
print(a)
```

### Variables

There are different types of variables
1. Numbers
2. Strings
3. Tuples
4. Lists
5. Dictionnaries

As indicated by its name, a number variable contains a number that can be of different "sub-types":
integer, long, float, comples, etc.

```python
n = 1
x = 2.23
print(n)
print(type(n))
print(type(x))
```

### String variables

A *string* variable contains a single or a string of characters surrounded by quotes or double-quotes

```python
c = 'a'
string = 'Hello! how are you?'
print(string)
string2 = ' today'
print(string+string2)
print(string*2)
```

### Tuples

A *tuple* contains a collection of elements of different types; it is immutable (cannot be changed once it is defined)

To define a tuple, use brackets ()

To access the different components of a tuple use indexing and square brackets []

```python
tup = (1, 2.34, 'Hi', c)
print(tup)
print(tup[1])
print(type(tup[3]))
```

### Lists

A *list* is like a tuple but it is mutable and is defined using square brackets []

Operations applying to strings (concatenation, multiplication, etc.) can also be applied to list

```python
lis = ['one', 23.5, 'a', 1]
print(lis + lis)
print(lis*3)
lis[1]='two'
print(lis)
```

### Dictionnaries

A Dictionary stores a collection of different data types elements in the form of key-value pairs.

Keys are unique

Dictionnaries mutable and are defined by using curly bracket {}

```python
dic = {'key1': [1,2,3], 'key2': 'John', 'key3': 2.345}
print(dic)
print(dic['key1'])
print(dic.keys())
print(dic.values())
print(dic.items())
```
