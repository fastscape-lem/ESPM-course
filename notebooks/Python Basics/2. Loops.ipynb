# Loops

To shorten the length of the code, and perform repetitive operations, one uses *loops*

A typical loop starts with a "for" statement that must end with a column ":"

The indentation indicates which lines are to be iterated in the loop

```python
lst = ['a', 1, '33.4']
for it in lst:
    print(it)
```

### Range

One can also iterate through a *range*

Note that, by default, a range starts at 0 and stops before the value indicated as the non-optional argument of range()

```python
for i in range(3):
    print(i)
```

A range can have up to three arguments (start, end, step)

```python
for i in range(1,3):
    print(i)
```

```python
for i in range (0,10,2):
    print(i)
```

```python
for i in range (10,0,-4):
    print(i)
```

### Enumerate

*enumerate* is an efficient way to go through a list of elements AND create an index

```python
lst = ['a', 1, '33.4']
for i, it in enumerate(lst):
    print(i, it)
```

### While loops

Another type of loop is a *while* loop that is performed as long as a condition is met

Conditional operators include:
1. ">" greater than
2. "<" smaller than
3. "==" equal to
4. "!=" not equal to
5. ">=" greater than or equal to
6. "<=" smaller than or equal to

```python
i=0
while i<3:
    print(i)
    i += 1
```

```python
i=0
while i<=3:
    print(i)
    i += 1
```

```python
i=0
while i!=3:
    print(i)
    i += 1
```

Exercice: write a loop that prints side-by-side the elements of an array in backward order from the one before last by step of 2 AND the index of the loop

```python
a = ['0','1','2','3','4','5']
for i,j in enumerate(range(len(a)-2,-1,-2)):
    print(a[j],i)
```
