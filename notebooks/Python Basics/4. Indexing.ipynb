# Indexing

*Indexing* accesses part of a list as indicated by a *range*

A *range* contains, by default, all elements of the list

But it can be tuned by specifying a *start*, *end* and *step*

```python
lst = ['a', 'b', 'c', 'd', 'e', 'f']
```

### Single index

A single value correspnds to an index starting at 0 and ening at n-1 where n is the length of the list

```python
print(len(lst))
print(lst[0])
print(lst[3])
print(lst[5])
```

### Start and end

Give a start and an end by separating them by a column ":"

```python
print(lst[0:2])
```

```python
print(lst[1:4])
```

```python
print(lst[3:])
```

### Negative index

A negative index "-i" refers to the ith element before the last one 

```python
print(lst[-1])
```

```python
print(lst[:-1])
```

### Step or strand

A third argument (separated by a column ":") can be used to indicate a step (or strand)

```python
print(lst[0:-1:2])
```

### Reverse order

To access a list in reverse order:

```python
print(lst[::-1])
```

Exercice: print the elements of an array in reverse order starting from the one before last

```python
print(lst[-2::-1])
```
